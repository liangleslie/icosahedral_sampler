# Icosahedral sampler
---

```python
import cv2
import matplotlib.pyplot as plt

from ico_sampler import IcosahedralSampler
```

```python
ico_sampler = IcosahedralSampler(resolution=800)

print("Vertices: ", ico_sampler.vertices.shape)
print("Faces:    ", ico_sampler.faces.shape)
```

```python
image = cv2.imread('./assets/0.png')

plt.imshow(image[:, :, ::-1])
```

## Unwrap icosahedron
---

```python
icosa_image = ico_sampler.unwrap(image, face_offset=0)

plt.imshow(icosa_image[:, :, ::-1])
# cv2.imwrite('output.png', icosa_image)
```

```python
# face offset in interval [-2, 2]
for i in range(-2, 3):
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    icosa_image = ico_sampler.unwrap(image, face_offset=i)
    ax.imshow(icosa_image[:, :, ::-1])
    ax.set_title(f"Face offset: {i}")
```

## Face image
---

```python
# get face projection from the equirectangular image
# faces in range [0, 19]
face_image = ico_sampler.get_face_image(face_no=10, eq_image=image)

plt.imshow(face_image[:, :, ::-1])
```
